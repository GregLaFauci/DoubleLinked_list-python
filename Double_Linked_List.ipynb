```python
# Node class
class Node:
    def __init__(self, item, next_link = None, prev_link = None):
        self.item = item
        self.next_link = next_link # 다음 노드를 가리키는 링크
        self.prev_link = prev_link # 이전 노드를 가리키는 링크
```

```python
# Double Linked List class
class DList:
    def __init__(self):
        self.head = Node(None)
        self.tail = self.head
        self.size = 0
    
    def Is_empty(self): return self.size == 0
    def Size(self): return self.size
    
    def Empty_list_insert(self, item): # 빈 리스트일 경우 삽입하는 메소드
        self.head = Node(item)
        self.tail = self.head
        self.size += 1
    ### Node 삽입 
    def Insert_head(self, item): # 리스트의 첫번째에 노드 삽입
        if self.Is_empty():
            self.Empty_list_insert(item)
        else:
            current = self.head
            self.head = Node(item,current)
            current.prev_link = self.head
            self.size += 1
    def Insert_at_index(self, item, index): # X번째 자리에 노드 삽입
        current = self.head
        current_index = 0
        if index == 0: # 0번째 자리에 삽입은 head에 삽입하는것과 같으므로 insert_head()실행
            self.Insert_head(item)
        elif index == self.size: # insert_tail을 이용하지 않고 리스트의 맨 끝에 삽입하는 경우
            self.Insert_tail(item)
        else:
            while current_index < self.size-1:
                if current_index == index-1:
                    break;
                current = current.next_link
                current_index += 1
            new_node = Node(item,current.next_link,current)
            current.next_link.prev_link = new_node
            current.next_link = new_node
            self.size += 1
    def Insert_tail(self, item): # 리스트의 마지막에 노드 삽입
        current = self.tail
        self.tail = Node(item, None, current)
        current.next_link = self.tail
        self.size += 1
    
    ### Node 삭제
    def Delete_head(self): # 리스트의 첫번째 노드 삭제
        current = self.head
        self.head = current.next_link
        current.next_link = None
        self.head.prev_link = None
        self.size -= 1
    def Delete_at_index(self, index): # X번째 자리의 노드 삭제
        current = self.head
        current_index = 0
        if index == 0: # 리스트의 첫번째 노드를 삭제하는 경우
            self.Delete_head()
        else:
            while current_index < self.size-1:
                if current_index == index-1:
                    break;
                    current = current.next_link
                    current_index += 1
            delete_node = current.next_link
            delete_node.next_link.prev_link = current
            current.next_link = delete_node.next_link
            self.size -= 1
        """ 삭제할 노드의 link를 delete하기 위한 코드로 꼭 필요한것은 아님 
        delete_node.next_link = None
        delete_node.prev_link = None
        """
    def Delete_tail(self): # 리스트의 마지막 노드 삭제
        current = self.tail
        self.tail = current.prev_link
        self.tail.next_link = None
        current.prev_link = None
        self.size -= 1
    
    ### 특정 값을 가진 Node의 index number return
    def Search(self, target):
        current = self.head
        for i in range(self.size):
            if target == current.item:
                return i
            current = current.next_link
        return None
        
        
    ### 리스트 출력
    def Print_list(self): # 리스트의 첫번째 노드부터 출력
        current = self.head
        while current:
            if current.next_link != None:
                print(current.item, ' -> ', end = '')
            else:
                print(current.item)
            current = current.next_link
    def Print_list_back(self): # 리스트의 마지막 노드부터 출력
        current = self.tail
        while current:
            if current.prev_link != None:
                print(current.item, ' -> ', end = '')
            else:
                print(current.item)
            current = current.prev_link
```

```python
d = DList()
d.Insert_head(1)
print("현재 사이즈 : ",d.size)
d.Print_list()
d.Print_list_back()
print("-----------")

d.Insert_at_index(2, 0)
print("현재 사이즈 : ",d.size)
d.Print_list()
d.Print_list_back()
print("-----------")

d.Insert_tail(6)
print("현재 사이즈 : ",d.size)
d.Print_list()
d.Print_list_back()
print("-----------")

d.Insert_at_index(3, 1)
print("현재 사이즈 : ",d.size)
d.Print_list()
d.Print_list_back()
print("-----------")

d.Insert_tail(5)
print("현재 사이즈 : ",d.size)
d.Print_list()
d.Print_list_back()
print("-----------")

d.Delete_tail()
print("현재 사이즈 : ",d.size)
d.Print_list()
d.Print_list_back()
print("-----------")

d.Delete_at_index(1)
print("현재 사이즈 : ",d.size)
d.Print_list()
d.Print_list_back()
print("-----------")

d.Delete_at_index(0)
print("현재 사이즈 : ",d.size)
d.Print_list()
d.Print_list_back()
print("-----------")

d.Insert_at_index(4,2)
print("현재 사이즈 : ",d.size)
d.Print_list()
d.Print_list_back()
print("-----------")

d.Insert_at_index(9,2)
print("현재 사이즈 : ",d.size)
d.Print_list()
d.Print_list_back()
print("-----------")
```
